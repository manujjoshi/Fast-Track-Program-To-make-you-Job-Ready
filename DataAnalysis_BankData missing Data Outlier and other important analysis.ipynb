```python
### Removal of ? mark in Salary column

### Method - I
salary = []
for val in df['salary']:
    if '?' in val:
        salary.append(val.replace('?','nan'))
    else:
        salary.append(val)
```

```python
### Method = II
#df['salary'] = df['salary'].apply(lambda x:x.replace('?','nan'))
```

```python
import re
```

```python
### Method - III

##df['salary'].apply(lambda x:re.sub(r'\W','nan',x))
```

```python
### Method - IV
df['salary'] = df['salary'].replace(to_replace='?',value = np.nan)
```

```python
df['salary'].dtypes
```

```python
### Conversion of object salary column to numerical
## Method - I
df['salary'].astype('float')

## Method - II
pd.to_numeric(df['salary'])

### Method - III
df['salary'] = df['salary'].apply(lambda x:float(x))
```
